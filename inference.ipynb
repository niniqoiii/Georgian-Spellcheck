# Assignment 2 — Georgian Spellchecker: Inference

This notebook loads the trained model and demonstrates spelling correction on Georgian words.


```python
from spellchecker import correct_word, download_words_from_url
```

## Dataset refresh

The project uses the Google Drive dataset link. When internet is available, this cell downloads the latest version into `data/georgian_words.txt`. If internet is not available, the cached local file is used.


```python
DATASET_URL = 'https://drive.google.com/file/d/1NnELSMHpI9ru6RgzGVT6wiIiTAjWKgj_/view?usp=sharing'
try:
    download_words_from_url(DATASET_URL, 'data/georgian_words.txt')
    print('Dataset refreshed from Google Drive.')
except Exception as e:
    print('Could not download from Drive; using cached local dataset instead.')
    print(type(e).__name__, e)
```

## Required function

The assignment requires this exact function signature.


```python
def correct_word_required(word: str, model_path: str) -> str:
    """
    Takes a potentially misspelled Georgian word and returns the corrected version.
    """
    return correct_word(word, model_path)
```

## Demonstration

The examples include both misspelled and already-correct Georgian words.

```python
model_path = 'model/georgian_spellchecker.pt'
examples = [
    'გამარჰონა', 'გამარჯობა',
    'პროგამა', 'პროგრამა',
    'თბილსი', 'თბილისი',
    'საქართელო', 'საქართველო',
    'კომპიუტრი', 'კომპიუტერი',
    'სტუდნტი', 'სტუდენტი',
    'უნივერსტეტი', 'უნივერსიტეტი',
    'მეგობრი', 'მეგობარი',
    'სიტყა', 'სიტყვა',
    'შეცდმა', 'შეცდომა',
    'დავალბა', 'დავალება',
    'მოდლი', 'მოდელი'
]

for word in examples:
    print(f'{word} → {correct_word_required(word, model_path)}')
```
